```python
# Mengimpor library yang diperlukan
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Mengimpor dataset dari URL
url = '/content/drive/MyDrive/Machine Learning/diamonds.csv'
diamonds = pd.read_csv(url)

# Menampilkan lima baris pertama dari dataset
print(diamonds.head())

# Visualisasi scatter plot harga berdasarkan berat karat
plt.figure(figsize=(10, 6))
sns.scatterplot(data=diamonds, x='carat', y='price', alpha=0.5)
plt.title('Scatter Plot Harga Berdasarkan Berat Karat')
plt.xlabel('Berat Karat')
plt.ylabel('Harga')
plt.show()

# Visualisasi histogram dari berat karat
plt.figure(figsize=(10, 6))
sns.histplot(diamonds['carat'], kde=True, bins=30)
plt.title('Histogram Berat Karat')
plt.xlabel('Berat Karat')
plt.ylabel('Frekuensi')
plt.show()

# Visualisasi boxplot harga berdasarkan jenis potongan (cut)
plt.figure(figsize=(10, 6))
sns.boxplot(data=diamonds, x='cut', y='price')
plt.title('Boxplot Harga Berdasarkan Jenis Potongan (Cut)')
plt.xlabel('Cut')
plt.ylabel('Harga')
plt.show()

# Visualisasi distribusi harga
plt.figure(figsize=(10, 6))
sns.histplot(diamonds['price'], kde=True, bins=30)
plt.title('Distribusi Harga')
plt.xlabel('Harga')
plt.ylabel('Frekuensi')
plt.show()

# Visualisasi distribusi harga berdasarkan jenis potongan (cut)
plt.figure(figsize=(10, 6))
sns.violinplot(data=diamonds, x='cut', y='price')
plt.title('Distribusi Harga Berdasarkan Jenis Potongan (Cut)')
plt.xlabel('Cut')
plt.ylabel('Harga')
plt.show()
```
